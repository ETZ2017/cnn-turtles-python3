```python
result = data[data['label'] == 'Certain Turtle'][data['filename'].isin(data[data['label'].isin(['0', '1', '3', '4'])]['filename'].unique())]
len(result)
```

```python
repeated = result[0:1][0:1].values[0][1]
# repeated = result.sample(n=1, random_state=42).values[0][1]
```

```python
sample = data[(data['filename'] == repeated) & (data['label'].isin(['Certain Turtle', '0', '1', '3', '4']))].sort_values(by=['label'])
sample
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.patches as patches

# Create a 2x3 grid of subplots (2 rows, 3 columns)
fig, axs = plt.subplots(3, 4, figsize=(20, 12))  # Adjust the figsize as needed

# Flatten the 2D array of subplots into a 1D array
axs = axs.flatten()

for i in range(len(sample[0:12])):

    folder_name = sample[i:i+1].values[0][0]
    file_name = sample[i:i+1].values[0][1]
    
    x = sample[i:i+1].values[0][5]
    y = sample[i:i+1].values[0][4]

    # Load the image
    img = mpimg.imread(f'../{folder_name}/{file_name}') 

    # Display the image
    axs[i].imshow(img)
    circle = patches.Circle((x,y), 200, edgecolor='r', facecolor='None')  # You can customize edgecolor and facecolor

    # Add the circle to the plot
    axs[i].add_patch(circle)

# Show the plot
plt.show()
```
